```
# Print all available voices
import pyttsx3
engine = pyttsx3.init()

voices = engine.getProperty('voices')
for voice in voices:
    print("Voice:")
    print(" - ID: %s" % voice.id)
    print(" - Name: %s" % voice.name)
    print(" - Languages: %s" % voice.languages)
    print(" - Gender: %s" % voice.gender)
    print(" - Age: %s" % voice.age)
```

```
import pyttsx3
engine = pyttsx3.init(driverName='espeak')

# Voice IDs pulled from engine.getProperty('voices')
# These will be system specific
bu_voice_id = "english-us"

# Use female English voice
engine.setProperty('voice', bu_voice_id)
Us_Female = engine.say('Hello with my new voice, what do you think?')

def save_to_file(self, text, filename, name=None):
 
    self.proxy.save_to_file(text, filename, name)

engine.save_to_file(Us_Female, 'voicy')

engine.runAndWait()
engine.stop()
```
